#import the dependecies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

```

#Data collection and Preprocessing

```python
#loading the dataset to pandas dataframe
loan_dataset = pd.read_csv('loan_prediction.csv')
```

```python
#printing the 1st 5 rows of the dataframe
loan_dataset.head()
```

```python
#print rows and column
loan_dataset.shape
```

```python
#statistical measure
loan_dataset.describe()
```

```python
#number of missing values in each column
loan_dataset.isnull().sum()
```

```python
#removing/dropping the missing value
loan_dataset = loan_dataset.dropna()
```

```python
loan_dataset.isnull().sum()
```

```python
# label encoding & printing value
loan_dataset.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
loan_dataset.head()
```

```python
#Dependent column value : means 1 dependet ->1 , 2 dependent ->2 values
loan_dataset['Dependents'].value_counts()
```

```python
#replacing 3+ to 4
loan_dataset.replace({'Dependents':{'3+':4}},inplace=True)
```

```python
loan_dataset['Dependents'].value_counts()
```

#Data Visulaization

```python
#education and loan status
sns.countplot(x='Education', hue='Loan_Status',data=loan_dataset)
```

```python
sns.barplot(x='Education', y='Loan_Status', hue='Dependents', data=loan_dataset)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(loan_dataset.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

```python
#marital status and loan status
sns.countplot(x='Married',hue='Loan_Status',data=loan_dataset)
```

```python
#label encoding
le = LabelEncoder()
loan_dataset['Married'] = le.fit_transform(loan_dataset['Married'])
loan_dataset['Gender'] = le.fit_transform(loan_dataset['Gender'])
loan_dataset['Property_Area'] = le.fit_transform(loan_dataset['Property_Area'])
loan_dataset['Education'] = le.fit_transform(loan_dataset['Education'])
loan_dataset.replace({'Self_Employed':{'No':0, 'Yes':1}},inplace=True)
```

```python
loan_dataset.head()
```

```python
#separating the data and label
X = loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis = 1)
Y = loan_dataset['Loan_Status']
```

```python
print(X)
print("\n")
print(Y)
```

#Feature Scaling : Using Standard_Scaler

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

#Split into Training and Test Data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.1,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

#Training the model:
##Support Vector Machine

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine
classifier.fit(X_train,Y_train)
```

#Model Evaluation:

```python
# accuracy_score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data:",training_data_accuracy)
```

```python
# accuracy_score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Test data: ",test_data_accuracy)
```
